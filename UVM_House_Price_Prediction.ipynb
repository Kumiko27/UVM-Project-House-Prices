```python
def create_regplot(df, x, y, xlbl, ylbl):
    plt.figure(figsize=(8,4), dpi=200)
    with sns.axes_style('darkgrid'):
        ax = sns.regplot(data=df,
                         x=x,
                         y=y,
                         color= '#2f4b7c',
                         scatter_kws={'alpha': 0.3},
                         line_kws={'color':'#ff7c43'})
        ax.set(ylabel=ylbl,
               xlabel=xlbl)
        plt.show()
```

```python
stand_cols =  ['StreetLineLength', 'ParcelSize', 'BasementFinishedArea1', 'BasementFinishedArea2',
       'BasementUnfinishedArea', 'LowQualityArea', 'LivingArea', 'BedroomAbvGr', 'TotalRooms', 'GarageCapacity',
       'GarageSize', 'WoodDeckArea', 'OpenPorchArea', 'EnclosedPorchArea', 'ThreeSeasonPorchArea', 'ScreenPorchArea',
       'House_Age', 'AgeOfGarage', 'TotalArea', 'TotalBathrooms']

hotenc_cols = ['BuildingCategory','ZoningClassification','MaterialQuality','ParcelShape','ConditionRating',
               'ParcelSettings', 'District', 'DwellingType', 'DwellingStyle', 'RoofType','ExteriorCladding1',
               'ExteriorCladding2','FoundationType', 'GarageLocation', 'SaleCondition']

ordenc_cols =['ExteriorQuality','BasementHeight','BasementFinish1', 'BasementAccess', 'HeatingQuality', 'KitchenQuality', 'GarageInterior']

mapper = [
    # 'ExteriorQuality',
    ['Fa', 'TA', 'Gd',  'Ex'],
    # 'BasementHeight',
    ['No', 'Fa', 'TA', 'Gd',  'Ex'],
    # 'BasementFinish1',
    ['No', 'Unf', 'LwQ', 'BLQ', 'Rec', 'ALQ', 'GLQ'],
    # 'BasementAccess',
    ['No', 'Mn', 'Av', 'Gd'],
    # 'HeatingQuality',
    ['Po', 'Fa', 'TA', 'Gd',  'Ex'],
    # 'KitchenQuality',
    ['Fa', 'TA', 'Gd',  'Ex'],
    # 'GarageInterior',
    ['No', 'Unf', 'RFn', 'Fin']
]
```

### 4.2 Normalizar (Escalar) datos de entrenamiento

**IMPORTANTE**: Ajustar la normalizaci√≥n en los datos de entrenamiento, NO en los datos de prueba.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
import scipy.stats as stats
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

#### 4.2.1 Normalizaci√≥n

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

preprocessor = ColumnTransformer(
    transformers=[
        ('one_hot', OneHotEncoder(handle_unknown='ignore'), hotenc_cols),
        ('ordinal', OrdinalEncoder(categories=mapper), ordenc_cols),
        ('num', StandardScaler(), stand_cols)
    ],
)
```

```python
X_train = preprocessor.fit_transform(X_train)
```

```python
X_test = preprocessor.transform(X_test)
```
